4. What is the total military expenditure as a percentage of GDP for all countries?

```python
# Calculate the total military expenditure and total GDP
total_military_expenditure = df['military_expenditure_pct_gdp'].sum()
total_gdp = df['gdp'].sum()

# Calculate the percentage of military expenditure to GDP for each country
df['military_expenditure_pct_of_gdp'] = (df['military_expenditure_pct_gdp'] / 100) * df['gdp']

# Create a DataFrame with country names and their military expenditure percentages
military_expenditure_df = df[['country', 'military_expenditure_pct_of_gdp']]

# Sort the DataFrame by military expenditure percentage in descending order
military_expenditure_df_sorted = military_expenditure_df.sort_values(by='military_expenditure_pct_of_gdp', ascending=False)

# Get the top 10 countries with the highest military expenditure as a percentage of GDP
high_military_expenditure = military_expenditure_df_sorted.head(10)

# Print the result
print(high_military_expenditure)
```

5. Which region has the highest average life expectancy?

```python
# Finding the average life expectancy for each region
average_life_expectancy_by_region = df.groupby('region')['life_expectancy'].mean()

# Arranging the regions by average life expectancy in descending order
sorted_regions = average_life_expectancy_by_region.sort_values(ascending=False)

# Get the top 10 regions with the highest average life expectancy
top_10_regions = sorted_regions.head(10)

# Calculate the count and percentage for the top 10 regions
total_regions = len(df['region'].unique())
top_10_count = len(top_10_regions)
top_10_percentage = (top_10_count / total_regions) * 100

# Print the top 10 regions with count and percentage
print("Top 10 Regions with the Highest Average Life Expectancy:")
print(top_10_regions)
```

```python
region_avg_life_expectancy = df.groupby('region')['life_expectancy'].mean().reset_index()

# Sort the regions by average life expectancy in descending order
region_avg_life_expectancy_sorted = region_avg_life_expectancy.sort_values(by='life_expectancy', ascending=False)

# Create a bar graph to visualize the regions with the highest average life expectancy
plt.figure(figsize=(12, 8))
sns.barplot(data=region_avg_life_expectancy_sorted, x='region', y='life_expectancy')
plt.xlabel('Region')
plt.ylabel('Average Life Expectancy')
plt.title('Average Life Expectancy by Region')
plt.xticks(rotation=90)
plt.show()
```

Analysis:-

Average life Expectency of All the Region are as Shown in Garph, Where Life Expectency is much higher in Australia and New Zealand region

6. Calculating the total no of gdp and which country contributes how much percentage in world gdp

```python
# Calculate the total world GDP
total_world_gdp = df['gdp'].sum()

# Calculate the GDP percentage contribution for each country
df['gdp_percentage_contribution'] = (df['gdp'] / total_world_gdp) * 100

# Create a DataFrame with country names and their GDP percentage contributions
gdp_contributions_df = df[['country', 'gdp_percentage_contribution']]

# Sort the DataFrame by GDP percentage contribution in descending order
gdp_contributions_sorted = gdp_contributions_df.sort_values(by='gdp_percentage_contribution', ascending=False)

# Print the total world GDP
print(f"Total World GDP: {total_world_gdp:.2f}")

# Print the top 10 countries with their GDP percentage contributions
top_10_gdp_contributions = gdp_contributions_sorted.head(10)
print("\nTop 10 Countries by GDP Percentage Contribution:")
print(top_10_gdp_contributions)
```

```python
# Extract the top 10 countries and their GDP percentage contributions
top_10_countries = top_10_gdp_contributions['country']
top_10_contributions = top_10_gdp_contributions['gdp_percentage_contribution']

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_10_contributions, labels=top_10_countries, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Countries by GDP Percentage Contribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```

Analysis

Here The United States of America is the highest contributer in world GDP

7) To calculate and visualize the average CO2 emissions per capita for each continent and identify the continent with the highest and lowest average CO2 emissions per capita

```python
# Calculate the average CO2 emissions per capita for each continent
average_co2_per_capita_by_continent = df.groupby('continent')['co2_emissions'].mean()

# Find the continent with the highest average CO2 emissions per capita
highest_avg_co2_continent = average_co2_per_capita_by_continent.idxmax()
highest_avg_co2_value = average_co2_per_capita_by_continent.max()

# Find the continent with the lowest average CO2 emissions per capita
lowest_avg_co2_continent = average_co2_per_capita_by_continent.idxmin()
lowest_avg_co2_value = average_co2_per_capita_by_continent.min()

# Print the results
print("Average CO2 Emissions Per Capita by Continent:")
print(average_co2_per_capita_by_continent)
print(f"Continent with Highest Average CO2 Emissions Per Capita: {highest_avg_co2_continent} ({highest_avg_co2_value:.2f} CO2 per capita)")
print(f"Continent with Lowest Average CO2 Emissions Per Capita: {lowest_avg_co2_continent} ({lowest_avg_co2_value:.2f} CO2 per capita)")
```

```python
# Calculate the average CO2 emissions per capita for each continent
average_co2_emissions = df.groupby('continent')['co2_emissions'].mean()

# Sort the continents by average CO2 emissions in descending order
sorted_continents = average_co2_emissions.sort_values(ascending=False)

# Get the continent with the highest average CO2 emissions per capita
highest_emissions_continent = sorted_continents.index[0]

# Get the continent with the lowest average CO2 emissions per capita
lowest_emissions_continent = sorted_continents.index[-1]

# Create a bar chart to visualize the average CO2 emissions per capita for each continent
plt.figure(figsize=(10, 6))
average_co2_emissions.plot(kind='barh', color='blue')
plt.xlabel('Continent')
plt.ylabel('Average CO2 Emissions per Capita')
plt.title('Average CO2 Emissions per Capita by Continent')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the bar chart
plt.show()

# Print the continent with the highest and lowest average CO2 emissions per capita
print('Anaylysis')
print(f"The continent with the highest average CO2 emissions per capita is {highest_emissions_continent}.")
print(f"The continent with the lowest average CO2 emissions per capita is {lowest_emissions_continent}.")
```

8) Finding the relation between GDP and Life Expectence 

```python
# Extract GDP and life expectancy data
gdp_data = df['gdp']
life_expectancy_data = df['life_expectancy']

# Create a scatter plot with GDP on the y-axis and life expectancy on the x-axis
plt.figure(figsize=(10, 6))
plt.scatter(life_expectancy_data, gdp_data, alpha=0.6)
plt.title('Scatter Plot: Life Expectancy vs. GDP')
plt.xlabel('Life Expectancy')
plt.ylabel('GDP')
plt.grid(True)

# Show the scatter plot
plt.show()

# Calculate the correlation between Life Expectancy and GDP
correlation = life_expectancy_data.corr(gdp_data)

# Print the correlation coefficient
print(f"Correlation between Life Expectancy and GDP: {correlation:.2f}")
```

Analysis

Countries with High GDP has the High Life Expectency and Countries with low GDP has Low Life Expectency

9) Comparing the population of male and female from each country?

```python
# Extract data for population of males and females
population_male = df['population_male']
population_female = df['population_female']
countries = df['country']

# Create an array of indices for the countries
x = np.arange(len(countries))

# Define the width of each bar
bar_width = 0.35

# Create the bar graph
plt.figure(figsize=(25, 8))
plt.bar(x - bar_width/2, population_male, bar_width, label='Male Population', alpha=0.7)
plt.bar(x + bar_width/2, population_female, bar_width, label='Female Population', alpha=0.7)

# Set labels, title, and ticks
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Comparison by Gender in Each Country')
plt.xticks(x, countries, rotation=90)
plt.legend()

# Show the bar graph
plt.tight_layout()
plt.show()
```

Analysis

China and India are the most populated country where both of these countries have more mens than womans
